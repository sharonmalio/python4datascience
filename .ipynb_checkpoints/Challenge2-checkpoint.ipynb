```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib as mpl
import matplotlib.pyplot as plt
%matplotlib inline
# Blood donations Dataset
test_data= pd.read_csv('BloodDonationPredictions/TestDataBlood.csv')
training_data = pd.read_csv('BloodDonationPredictions/TrainingDataBlood.csv')
# displaying the test_data table contents
table1= pd.DataFrame(test_data)
table1

```

```python
training_data = pd.read_csv('BloodDonationPredictions/TrainingDataBlood.csv')
table2= pd.DataFrame(training_data)
table2
```

```python
# number of donations against total volume donated
donations_volume = pd.DataFrame(test_data['Number of Donations'])
donations_volume.reset_index(inplace=True)
donations_volume.columns = ['Number of Donations', 'Total Volume Donated (c.c.)']

donations_volume=pd.DataFrame(donations_volume)
donations_volume
plt.figure(figsize=(20, 7))

barplot = sns.barplot(x=donations_volume['Number of Donations'], y=donations_volume['Total Volume Donated (c.c.)'])
barplot.set(xlabel='Number of Donations', ylabel='Total Volume Donated (c.c.)')
plt.title('Number of donations against the volume donated per person:', fontsize=20)
plt.xticks(fontsize=17)
plt.yticks(fontsize=17)
# plt.legend()
plt.show()
```

```python
# months since first donation number of donations
month_since_donation = pd.DataFrame(test_data['Number of Donations'])
month_since_donation.reset_index(inplace=True)
month_since_donation.columns = ['Months since First Donation', 'Number of Donations']

month_since_donation=pd.DataFrame(donations_volume)
month_since_donation
plt.figure(figsize=(20, 7))

barplot = sns.barplot(x=month_since_donation['Number of Donations'], y=month_since_donation['Months since First Donation'])
barplot.set(xlabel='Number of Donations', ylabel='Months since First Donation')
plt.title('Number of donations against the volume donated per person:', fontsize=20)
plt.xticks(fontsize=17)
plt.yticks(fontsize=17)
plt.legend()
plt.show()
```
